# Variáveis gerais

```python
def calcula_massa(velocidade = None, energia_cinetica = None, energia_potencial_gravitacional = None,
                  aceleracao_gravitacional = None, altura = None):
  """Calcula a massa a partir de velocidade e energia cinética ou energia potencial gravitacional, aceleração gravitacional e altura.

    Argumentos:
        velocidade (None ou float): Velocidade em metros por segundo (m/s)
            (padrão é None)
        energia_cinetica (None ou float): Energia cinética em joule (J)
            (padrão é None)
        energia_potencial_gravitacional (None ou float): Energia potencial gravitacional em joule (J)
            (padrão é None)
        aceleracao_gravitacional (None ou float): Aceleração gravitacional em metros por segundo ao quadrado (m/s**2)
            (padrão é None)
        altura (None ou float): Altura em metros (m)
            (padrão é None)
        densidade (None ou float): Densidade em quilogramas por metro cúbico (kg/m**3)
            (padrão é None)
        volume (None ou float): Volume em metros cúbicos (m**3)
            (padrão é None)

    Retornos:
        float: massa em quilogramas (kg)
  """

  massa = 0

  if velocidade != None and energia_cinetica != None:
    massa = 2*energia_cinetica/velocidade**2

  elif energia_potencial_gravitacional != None and aceleracao_gravitacional != None and altura != None:
    massa = energia_potencial_gravitacional/(aceleracao_gravitacional*altura)

  return massa

def calcula_velocidade(massa, energia_cinetica):
  """Calcula a velocidade a partir da massa e energia cinética.

    Argumentos:
        massa (None ou float): Massa em quilogramas (kg)
        energia_cinetica (None ou float): Energia cinética em joule (J)

    Retornos:
        float: velocidade em metros por segundo (m/s)
  """

  velocidade = (2*energia_cinetica/massa)**0.5

  return velocidade

def calcula_altura(energia_potencial_gravitacional, massa, aceleracao_gravitacional):
  """Calcula a altura a partir da energia potencial gravitacional, massa e açeleração gravitacional

    Argumentos:
        energia_potencial_gravitacional (bool ou float): Energia potencial gravitacional em joule (J)
        massa (bool ou float): Massa em quilogramas (kg)
        aceleracao_gravitacional (bool ou float): aceleração gravitacional em metros por segundo ao quadrado (m/s**2)

    Retornos:
        float: altura em metros (m)
  """

  altura = energia_potencial_gravitacional/(aceleracao_gravitacional*massa)

  return altura
```

# Energia

```python
def calcula_energia_cinetica(massa = None, velocidade = None, energia_potencial = None, energia_mecanica = None):
  """Calcula a energia cinética a partir da massa e velocidade ou energia potencial e energia mecânica.

    Argumentos:
        massa (None ou float): Massa em quilogramas (kg)
            (padrão é None)
        velocidade (None ou float): Velocidade em metros por segundo (m/s)
            (padrão é None)
        energia_potencial (None ou float): Energia potencial em joule (J)
            (padrão é None)
        energia_mecanica (None ou float): Energia mecânica em joule (J)
            (padrão é None)

    Retornos:
        float: energia cinética em joule (J)
  """

  energia_cinetica = 0

  if massa != None and velocidade != None:
    energia_cinetica = (0.5)*massa*(velocidade)**2

  elif energia_mecanica != None and energia_potencial != None:
    energia_cinetica = energia_mecanica - energia_potencial

  return energia_cinetica


def calcula_energia_potencial_gravitacional(massa, aceleracao_gravitacional, altura):
  """Calcula a energia potencial gravitacional a partir da massa, aceleração gravitacional e altura

    Argumentos:
        massa (float): Massa em quilogramas (kg)
        aceleracao_gravitacional (bool ou float): Aceleração gravitacional em metros por segundo ao quadrado (m/s**2)
        altura (float): Altura em metros (m)

    Retornos:
        float: energia gravitacional em joule (J)
  """

  energia_potencial_gravitacional = massa*aceleracao_gravitacional*altura

  return energia_potencial_gravitacional


def calcula_aceleracao_gravitacional(energia_potencial_gravitacional, massa, altura):
  """Calcula a aceleração gravitacional a partir de energia potencial gravitacional, massa e altura

    Argumentos:
        energia_potencial_gravitacional (bool ou float): Energia potencial gravitacional em joule (J)
        massa (bool ou float): Massa em quilogramas (kg)
        altura (bool ou float): Altura em metros (m)

    Retornos:
        float: aceleração gravitacional em metros por segundo ao quadrado (m/s**2)
  """

  aceleracao_gravitacional = energia_potencial_gravitacional/(massa*altura)

  return aceleracao_gravitacional


def calcula_energia_mecanica(energia_cinetica, energia_potencial):
  """Calcula a energia mecânica a partir de energia cinética e energia potencial

    Argumentos:
        energia_cinetica (loat): Energia cinética em joule (J)
        energia_potencial (float): Energia potencial em joule (J)

    Retornos:
        float: energia mecânica em joule (J)
  """


  energia_mecanica = energia_cinetica + energia_potencial

  return energia_mecanica

def calcula_energia_potencial_elastica(constante_elastica, elongacao):
  """Calcula a energia potencial elástica a partir da constante elástica e elongação

    Argumentos:
        constante_elastica (float): Constante elástica em newton por metro (N/m)
        elongacao (float): Elongação em metros (m)

    Retornos:
        float: energia potencial elástica em joule (J)
  """

  energia_potencial_elastica = (0.5)*constante_elastica*(elongacao)**2

  return energia_potencial_elastica


def calcula_energia_potencial(energia_cinetica, energia_mecanica):
  """Calcula a energia potencial a partir de energia cinética e energia mecânica

    Argumentos:
        energia_cinetica (float): Energia cinética em joule (J)
        energia_mecanica (float): Energia mecânica em joule (J)

    Retornos:
        float: energia potencial em joule (J)
  """

  energia_potencial = energia_mecanica - energia_cinetica

  return energia_potencial
```

### Função Horária das Posições

```python
import pandas as pd
from matplotlib import pyplot as plt

def funcao_horaria_posicoes(posicao_inicial, velocidade_inicial, tempo, aceleracao):
  """Fornece a posição através da função horária das posições, partindo de posição inicial, velocidade inicial, tempo e aceleração.

    Argumentos:
        posicao_inicial (float): Posição inicial em metros (m)
        velocidade_inicial (float): Velocidade inicial em metros por segundo (m/s)
        tempo (float): Tempo em segundos (s)
        aceleração (float): Aceleração em metros por segundo ao quadrado (m/s**2)

    Retornos:
        float: posição em metros (m)
  """

  posicao = posicao_inicial + velocidade_inicial * tempo + (aceleracao * tempo**2)/2
  return posicao

def plotar_funcao_horaria_posicoes(tempo_final, tempo_inicial = 0, aceleracao = 0, posicao_inicial = 0, velocidade_inicial = 0, passo = 0):
  """Plota o gráfico da função horária das posições a partir dos parâmetros inseridos

    Argumentos:
        tempo_final (float): Tempo final em segundos (s)
        tempo_inicial (float): Tempo inicial em segundos (s)
            (padrão é 0)
        aceleracao (float): Aceleração em metros por segundo ao quadrado (m/s**2)
            (padrão é 0)
        posicao_inicial (float): Posição inicial em metros (m)
            (padrão é 0)
        velocidade_inicial (float): Velocidade inicial em metros por segundo (m/s)
            (padrão é 0)
        passo (float): Intervalo entre as medições em segundos (s)
            (padrão é 1)

    Retornos:
        matplotlib.axes._axes.Axes: gráfico da função horária das posições de posição (m) por tempo (s).
  """

  tempos = []
  posicoes = []
  data_frame = pd.DataFrame(columns =['Tempo (s)', 'Posição (m)'])
  for tempo in range (tempo_inicial, tempo_final, passo):
    tempos.append(tempo)
    posicoes.append(funcao_horaria_posicoes(posicao_inicial = posicao_inicial, velocidade_inicial = velocidade_inicial, tempo = tempo, aceleracao = aceleracao))
    tempo += passo
  data_frame['Tempo (s)'] = tempos
  data_frame['Posição (m)'] = posicoes

  data_frame.plot.scatter('Tempo (s)', 'Posição (m)', color = 'black')
  plt.show()
```

# Força elástica

```python
#f= força em newtons
#k= constante newtons/metro
#x= deformação em metros
#e= energia
#cf= comprimento final da mola
#ci= comprimento inical da mola

#é necessário fornecer o valor da constante elastica, pois esta muda de acordo com o material

from math import sqrt

def calcula_forca_elastica(k, x):
    """Calcula a força elástica a partir da constante elástica e da deformação da mola

    Argumentos:
        k (float): constante elástica (newtons/metro)
        x (float): deformação da mola (metros)

    Retornos:
        f (float): força elástica (newton)
    """
    f = k * x
    return f

def calcula_deformacao_da_mola (k = None, e = None, cf = None, ci = None, f = None):
    """Calcula a deformação da mola a partir da constante elástica e da força elástica ou da energia elástica e da constante elástica ou pela diferença de comprimento das molas

    Argumentos:
        k (float): constante elástica (newtons/metro)
          (padrâo é None)
        f (float): força elástica (newton)
          (padrâo é None)
        e (float): energia potencial elástica
          (padrâo é None)
        cf (float): comprimento final da mola
          (padrâo é None)
        ci (float): comprimento inicial da mola
          (padrâo é None)

    Retornos:
        x (float): deformação da mola (metros)
    """

    if f != None and k != None:
        x = f/k

    elif e!= None and k != None:
        x = sqrt((e*2)/k)

    elif cf !=None and ci != None:
        x = cf - ci

    return x

def calcula_constante_elastica(x=None, e=None, f=None):
    """Calcula a constante elástica a partir da deformação da mola e da força elástica ou da energia elástica e da deformação da mola

    Argumentos:
        x (float): deformação da mola (metros)
            (padrão é None)
        f (float): força elástica (newton)
            (padrão é None)
        e (float): energia elástica (J)
            (padrão é None)


    Retornos:
        k (float): constante da mola (newton/metros)
    """
    if f != None and x!= None:
        k = f/x
    elif e!= None and x!= None:
        k = (2*e)/x**2

    return k

def calcula_elongacao(constante_elastica = None, energia_potencial_elastica = None, forca_elastica = None, comprimento_inicial = None, comprimento_final = None):
  """Calcula a elongação a partir de constante elástica e energia potencial elástica ou constante elástica e força elástica ou comprimento inicial e comprimento final

    Argumentos:
        constante_elastica (None ou float): Constante elástica em newton por metro (N/m)
            (padrão é False)
        energia_potencial_elastica (None ou float): Energia potencial elástica em joule (J)
            (padrão é False)
        forca_elastica (None ou float): Força elástica em newton (N)
            (padrão é False)
        comprimento_inicial (None ou float): Comprimento inicial em metros (m)
            (padrão é False)
        comprimento_final (None ou float): Comprimento final em metros (m)
            (padrão é False)

    Retornos:
        float: elongação em metros (m)
  """

  elongacao = 0

  if constante_elastica != None and energia_potencial_elastica != None:
    elongacao = (2*energia_potencial_elastica/constante_elastica)**0.5

  elif forca_elastica != None and constante_elastica != None:
    elongacao = forca_elastica/constante_elastica

  elif comprimento_final !=None and comprimento_inicial != None:
    elongacao = comprimento_final - comprimento_inicial

  return elongacao
```

#Função horária das velocidades

```python
#V=Vo+at
#v: velocidade final
#vo:velocidade inicial
#a: aceleração
#t: tempo

import pandas as pd
import matplotlib.pyplot as plt



def velocidade_final (vo, a, t):
    """ A função velocidade_final, calcula o valor da velocidade final, tendo como dados a velocidade inicial, a aceleração e o tempo

    Argumentos:
        vo (int ou float): velocidade inicial
        a (int ou float): aceleração
        t (int ou float): tempo

    Return:
        A função retorna o valor numérico da velocidade final
    """
    v=vo+a*t
    return v

def dataframe(vo, a, t_inicial, t_final, passo):
    """ A função dataframe, define um DataFrame, que será utilizado para o gráfico de velocidade por tempo, tendo como dados o tempo e a velocidade final

    Argumentos:
        vo (int ou float): velocidade inicial
        a (int ou float): aceleração
        t_inicial (int): tempo inicial
        t_final (int): tempo final
        passo (int): intervalo entre os tempos

    Return:
        A função retorna um DataFrame de velocidade e tempo
    """
    t = [i for i in range(t_inicial, t_final + 1, passo)]
    dados= {'tempo(segundos)':t, 'velocidade(m/s)': [velocidade_final(vo, a, tempo) for tempo in t]}
    df = pd.DataFrame(dados)
    return df

def grafico (df):
    """ A função gráfico plota um gráfico a partir do DataFrame

    Argumentos:
        df (DataFrame): dataframe com os dados para o gráfico

    Return:
        A função retorna um gráfico
    """

    df = df.plot(x='tempo(segundos)', y='velocidade(m/s)')
    plt.xlabel('Tempo (segundos)')
    plt.ylabel('Velocidade (m/s)')
    plt.title('Gráfico de Velocidade x Tempo')
    plt.show()

def plotar_funcao_horaria_das_velocidades(vo, a, t_inicial, t_final, passo):
  df = dataframe(vo, a, t_inicial, t_final, passo)
  grafico(df)
```

# MRU

1ª Formula : S = So + V . t

```python
    # S = so + v .t

    # So = (v - t) / S

    # V = (So - t) / S

    # t = (So - v) /s
```

```python
def sorvete(sfinal=None, sinicial=None, velocidade=None, tempo=None):

    # S = so + v .t
    if sinicial != None and velocidade != None and tempo != None:
        sfinal = sinicial + velocidade * tempo
        return sfinal

    # So = (v - t) / S
    elif sfinal != None and velocidade != None and tempo != None:
        sinicial = sfinal - velocidade * tempo
        return sinicial

    # V = (So - t) / S
    elif sfinal != None and sinicial != None and tempo != None:
        velocidade = (sfinal - sinicial) / tempo
        return velocidade

    # t = (So - v) /s
    elif sfinal != None and sinicial != None and velocidade != None:
        tempo = (sfinal - sinicial) / velocidade
        return tempo

    else:
        return "Parâmetros insuficientes ou mal definidos"
```

2º Fórmula - Velocide Média

Vm = S - So / T - To

S = Vm . (T - To)+So

So = S - Vm . (t - To)

T = (S- So / Vm ) + To

```python
def velocidade_media(vmedia = None , sfinal = None, sinicial = None, tfinal = None, tinicial = None):

    # Vm = S - So / T - To

    if  sfinal != None and sinicial != None and tfinal != None and tinicial != None:
        vmedia = (sfinal - sinicial)/(tfinal-tinicial)
        return vmedia

    # S = Vm . (T - To)+So

    elif vmedia != None and sinicial != None and tfinal != None and tinicial != None:
        sfinal = vmedia * ( tfinal - tinicial) + sinicial
        return sfinal

    # So = S - Vm . (t - To)
    elif vmedia != None and sfinal != None and tfinal != None and tinicial != None:
        sinicial = sfinal - vmedia * (tfinal - tinicial)
        return sinicial

    # T = (S- So / Vm ) + To
    elif vmedia != None and sfinal != None and sinicial != None and tinicial != None:
        tfinal = ( sfinal - sinicial ) / vmedia + tinicial
        return tfinal

    # To = T - (S-So)/Vm
    elif vmedia != None and sfinal != None and sinicial != None and tfinal != None:
        tinicial = tfinal - (sfinal - sinicial)/vmedia
        return tinicial
```
