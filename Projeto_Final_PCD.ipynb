```python
def energia_cinetica(massa = False, velocidade = False, energia_potencial = False, energia_mecanica = False):
  """Calcula a energia cinética a partir da massa e velocidade ou energia potencial e energia mecânica.

    Argumentos:
        massa (bool ou float): Massa em quilogramas (kg)
            (padrão é False)
        velocidade (bool ou float): Velocidade em metros por segundo (m/s)
            (padrão é False)
        energia_potencial (bool ou float): Energia potencial em joule (J)
            (padrão é False)
        energia_mecanica (bool ou float): Energia mecânica em joule (J)
            (padrão é False)

    Retornos:
        float: energia cinética em joule (J)
  """

  if massa != False and velocidade != False:
    energia_cinetica = (0.5)*massa*(velocidade)**2

  elif energia_mecanica != False and energia_potencial != False:
    energia_cinetica = energia_mecanica - energia_potencial

  return energia_cinetica


def massa(velocidade = False, energia_cinetica = False, energia_potencial_gravitacional = False, aceleracao_gravitacional = False, altura = False):
  """Calcula a massa a partir de velocidade e energia cinética ou energia potencial gravitacional, aceleração gravitacional e altura.

    Argumentos:
        velocidade (bool ou float): Velocidade em metros por segundo (m/s)
            (padrão é False)
        energia_cinetica (bool ou float): Energia cinética em joule (J)
            (padrão é False)
        energia_potencial_gravitacional (bool ou float): Energia potencial gravitacional em joule (J)
            (padrão é False)
        aceleracao_gravitacional (bool ou float): Aceleração gravitacional em metros por segundo ao quadrado (m/s**2)
            (padrão é False)
        altura (bool ou float): Altura em metros (m)
            (padrão é False)

    Retornos:
        float: massa em quilogramas (kg)
  """

  if velocidade != False and energia_cinetica != False:
    massa = 2*energia_cinetica/velocidade**2

  elif energia_potencial_gravitacional != False and aceleracao_gravitacional != False and altura != False:
    massa = energia_potencial_gravitacional/(aceleracao_gravitacional*altura)

  return massa


def velocidade(massa = False, energia_cinetica = False):
  """Calcula a velocidade a partir da massa e energia cinética.

    Argumentos:
        massa (bool ou float): Massa em quilogramas (kg)
            (padrão é False)
        energia_cinetica (bool ou float): Energia cinética em joule (J)
            (padrão é False)

    Retornos:
        tipo: velocidade em metros por segundo (m/s)
  """

  if massa != False and energia_cinetica != False:
    velocidade = (2*energia_cinetica/massa)**0.5

  return velocidade


def energia_potencial_gravitacional(massa = False, aceleracao_gravitacional = False, altura = False):
  """Calcula a energia potencial gravitacional a partir da massa, aceleração gravitacional e altura

    Argumentos:
        massa (bool ou float): Massa em quilogramas (kg)
            (padrão é False)
        aceleracao_gravitacional (bool ou float): Aceleração gravitacional em metros por segundo ao quadrado (m/s**2)
            (padrão é False)
        altura (bool ou float): Altura em metros (m)
            (padrão é False)

    Retornos:
        float: energia gravitacional em joule (J)
  """

  if massa != False and aceleracao_gravitacional != False and altura != False:
    energia_potencial_gravitacional = massa*aceleracao_gravitacional*altura

  return energia_potencial_gravitacional


def altura(energia_potencial_gravitacional = False, massa = False, aceleracao_gravitacional = False):
  """Calcula a altura a partir da energia potencial gravitacional, massa e açeleração gravitacional

    Argumentos:
        energia_potencial_gravitacional (bool ou float): Energia potencial gravitacional em joule (J)
            (padrão é False)
        massa (bool ou float): Massa em quilogramas (kg)
            (padrão é False)
        aceleracao_gravitacional (bool ou float): aceleração gravitacional em metros por segundo ao quadrado (m/s**2)
            (padrão é False)

    Retornos:
        float: altura em metros (m)
  """

  if energia_potencial_gravitacional != False and massa != False and aceleracao_gravitacional != False:
    altura = energia_potencial_gravitacional/(aceleracao_gravitacional*massa)

  return altura


def aceleracao_gravitacional(energia_potencial_gravitacional = False, massa = False, altura = False):
  """Calcula a aceleração gravitacional a partir de energia potencial gravitacional, massa e altura

    Argumentos:
        energia_potencial_gravitacional (bool ou float): Energia potencial gravitacional em joule (J)
            (padrão é False)
        massa (bool ou float): Massa em quilogramas (kg)
            (padrão é False)
        altura (bool ou float): Altura em metros (m)
            (padrão é False)

    Retornos:
        float: aceleração gravitacional em metros por segundo ao quadrado (m/s**2)
  """

  if energia_potencial_gravitacional != False and massa != False and altura != False:
    aceleracao_gravitacional = energia_potencial_gravitacional/(massa*altura)

  return aceleracao_gravitacional


def energia_potencial_elastica(constante_elastica = False, elongacao = False):
  """Calcula a energia potencial elástica a partir da constante elástica e elongação

    Argumentos:
        constante_elastica (bool ou float): Constante elástica em newton por metro (N/m)
            (padrão é False)
        elongacao (bool ou float): Elongação em metros (m)
            (padrão é False)

    Retornos:
        float: energia potencial elástica em joule (J)
  """

  if constante_elastica != False and elongacao != False:
    energia_potencial_elastica = (0.5)*constante_elastica*(elongacao)**2
    return energia_potencial_elastica


def constante_elastica(forca_elastica = False, elongacao = False, energia_potencial_elastica = False):
  """Calcula a constante elástica a partir de força elástica e elongação ou energia potencial elástica e elongação

    Argumentos:
        forca_elastica (bool ou float): Força elástica em newton (N)
            (padrão é False)
        elongacao (bool ou float): Elongação em metros (m)
            (padrão é False)
        energia_potencial_elastica (bool ou float): Energia potencial elástica em joule (J)
            (padrão é False)

    Retornos:
        float: constante elástica em newton por metro (N/m)
  """

  if forca_elastica != False and elongacao != False:
    constante_elastica = forca_elastica/elongacao

  elif elongacao != False and energia_potencial_elastica != False:
    constante_elastica = 2*energia_potencial_elastica/elongacao**2

  return constante_elastica


def elongacao(constante_elastica = False, energia_potencial_elastica = False, forca_elastica = False, comprimento_inicial = False, comprimento_final = False):
  """Calcula a elongação a partir de constante elástica e energia potencial elástica ou constante elástica e força elástica ou comprimento inicial e comprimento final

    Argumentos:
        constante_elastica (bool ou float): Constante elástica em newton por metro (N/m)
            (padrão é False)
        energia_potencial_elastica (bool ou float): Energia potencial elástica em joule (J)
            (padrão é False)
        forca_elastica (bool ou float): Força elástica em newton (N)
            (padrão é False)
        comprimento_inicial (bool ou float): Comprimento inicial em metros (m)
            (padrão é False)
        comprimento_final (bool ou float): Comprimento final em metros (m)
            (padrão é False)

    Retornos:
        float: elongação em metros (m)
  """

  if constante_elastica != False and energia_potencial_elastica != False:
    elongacao = (2*energia_potencial_elastica/constante_elastica)**0.5

  if forca_elastica != False and constante_elastica != False:
    elongacao = forca_elastica/constante_elastica

  elif comprimento_final !=False and comprimento_inicial != False:
    elongacao = comprimento_final - comprimento_inicial

  return elongacao


def energia_mecanica(energia_cinetica = False, energia_potencial = False):
  """Calcula a energia mecânica a partir de energia cinética e energia potencial

    Argumentos:
        energia_cinetica (bool ou float): Energia cinética em joule (J)
            (padrão é False)
        energia_potencial (bool ou float): Energia potencial em joule (J)
            (padrão é False)

    Retornos:
        float: energia mecânica em joule (J)
  """

  if energia_mecanica == False and energia_potencial != False:
    energia_mecanica = energia_cinetica + energia_potencial

  return energia_mecanica


def energia_potencial(energia_cinetica = False, energia_mecanica = False):
  """Calcula a energia potencial a partir de energia cinética e energia mecânica

    Argumentos:
        energia_cinetica (bool ou float): Energia cinética em joule (J)
            (padrão é False)
        energia_mecanica (bool ou float): Energia mecânica em joule (J)
            (padrão é False)

    Retornos:
        float: energia potencial em joule (J)
  """

  if energia_mecanica != False and energia_cinetica != False:
    energia_potencial = energia_mecanica - energia_cinetica
#fazer um condicional para medir energia potencial se ela for só gravitacional e elástica somadas ou subtraídas?
  return energia_potencial

```

```python
import pandas as pd
from matplotlib import pyplot as plt

def funcao_horaria_posicoes(posicao_inicial, velocidade_inicial, tempo, aceleracao):
  """Fornece a posição através da função horária das posições, partindo de posição inicial, velocidade inicial, tempo e aceleração.

    Argumentos:
        posicao_inicial (float): Posição inicial em metros (m)
        velocidade_inicial (float): Velocidade inicial em metros por segundo (m/s)
        tempo (float): Tempo em segundos (s)
        aceleração (float): Aceleração em metros por segundo ao quadrado (m/s**2)

    Retornos:
        float: posição em metros (m)
  """

  posicao = posicao_inicial + velocidade_inicial * tempo + (aceleracao * tempo**2)/2
  return posicao

def plotar_funcao_horaria_posicoes(tempo_final, tempo_inicial = 0, aceleracao = 0, posicao_inicial = 0, velocidade_inicial = 0, passo = 1, erro_tempo = 0, erro_posicao = 0):
  """Plota o gráfico da função horária das posições a partir dos parâmetros inseridos

    Argumentos:
        tempo_final (float): Tempo final em segundos (s)
        tempo_inicial (float): Tempo inicial em segundos (s)
            (padrão é 0)
        aceleracao (float): Aceleração em metros por segundo ao quadrado (m/s**2)
            (padrão é 0)
        posicao_inicial (float): Posição inicial em metros (m)
            (padrão é 0)
        velocidade_inicial (float): Velocidade inicial em metros por segundo (m/s)
            (padrão é 0)
        passo (float): Intervalo entre as medições em segundos (s)
            (padrão é 1)
        erro_tempo (float): Erro do tempo em segundos (s)
            (padrão é 0)
        erro_posicao (float): Erro da posição em metros (m)
            (padrão é 0)

    Retornos:
        matplotlib.axes._axes.Axes: gráfico da função horária das posições de posição (m) por tempo (s).
  """

  tempos = []
  posicoes = []
  data_frame = pd.DataFrame(columns =['Tempo (s)', 'Posição (m)', 'Erro tempo (s)', 'Erro Posição (m)'])
  for tempo in range (tempo_inicial, tempo_final+passo):
    tempos.append(tempo)
    posicoes.append(funcao_horaria_posicoes(posicao_inicial = posicao_inicial, velocidade_inicial = velocidade_inicial, tempo = tempo, aceleracao = aceleracao))
    tempo += passo
  data_frame['Tempo (s)'] = tempos
  data_frame['Posição (m)'] = posicoes
  data_frame['Erro tempo (s)'] = erro_tempo
  data_frame['Erro Posição (m)'] = erro_posicao

  #x = data_frame["Tempo (s)"].values
  #y = data_frame["Posição (m)"].values
  #plt.errorbar('Tempo (s)', 'Posição (m)', xerr=data_frame["Erro Posição (m)"], yerr=data_frame["Erro Tempo (s)"], fmt = '.')
  return data_frame.plot.scatter('Tempo (s)', 'Posição (m)', color = 'black') #plt.show #plt.gca().spines[['top', 'right',]].set_visible(False)

plotar_funcao_horaria_posicoes(tempo_final = 100, tempo_inicial = 0, posicao_inicial = 0, aceleracao = 10, velocidade_inicial = 0, passo = 1, erro_tempo = 10, erro_posicao = 10)
```

# MRU

1ª Formula : S = So . V . t

```python
    # S = so + v .t

    # So = (v - t) / S

    # V = (So - t) / S

    # t = (So - v) /s
```

```python
def sorvete(sfinal=None, sinicial=None, velocidade=None, tempo=None):

    # S = so + v .t
    if sfinal == None and sinicial != None and velocidade != None and tempo != None:
        sfinal = sinicial + velocidade * tempo
        return sfinal

    # So = (v - t) / S
    elif sfinal != None and sinicial == None and velocidade != None and tempo != None:
        sinicial = sfinal - velocidade * tempo
        return sinicial

    # V = (So - t) / S
    elif sfinal != None and sinicial != None and velocidade == None and tempo != None:
        velocidade = (sfinal - sinicial) / tempo
        return velocidade

    # t = (So - v) /s
    elif sfinal != None and sinicial != None and velocidade != None and tempo == None:
        tempo = (sfinal - sinicial) / velocidade
        return tempo

    else:
        return "Parâmetros insuficientes ou mal definidos"
```

2º Fórmula - Velocide Média

Vm = S - So / T - To

S = Vm . (T - To)+So

So = S - Vm . (t - To)

T = (S- So / Vm ) + To

```python
def velocidade_media(vmedia = None , sfinal = None, sinicial = None, tfinal = None, tinicial = None, vfinal = None, vinicial = None):

    # Vm = S - So / T - To

    if  vmedia == None and sfinal != None and sinicial != None and tfinal != None and tinicial != None and vfinal == None and vinicial == None:
        vmedia = (sfinal - sinicial)/(tfinal-tinicial)
        return vmedia

    # S = Vm . (T - To)+So

    elif vmedia != None and sfinal == None and sinicial != None and tfinal != None and tinicial != None and vfinal != None and vinicial != None:
        sfinal = vmedia * ( tfinal - tinicial) + sinicial
        return sfinal

    # So = S - Vm . (t - To)
    elif vmedia != None and sfinal != None and sinicial == None and tfinal != None and tinicial != None and vfinal != None and vinicial != None:
        sinicial = sfinal - vmedia * (tfinal - tinicial)
        return sinicial

    # T = (S- So / Vm ) + To
    elif vmedia != None and sfinal != None and sinicial != None and tfinal == None and tinicial != None and vfinal != None and vinicial != None:
        tfinal = ( sfinal - sinicial ) / vmedia + tinicial
        return tfinal

    # To = T - (S-So)/Vm
    elif vmedia != None and sfinal != None and sinicial != None and tfinal != None and tinicial == None and vfinal != None and vinicial != None:
        tinicial = tfinal - (sfinal - sinicial)/vmedia
        return tinicial
```
